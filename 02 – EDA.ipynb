```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
ds = pd.read_csv("clean_data.csv")
ds.info()
```

```python
ds.info()
```

```python
company_count  = ds["company"].value_counts()
company_count
```

```python
fuel_type  = ds["fuel_type"].value_counts()
fuel_type
plt.pie(
    fuel_type,
    labels=fuel_type.index,
    autopct="%1.1f%%",
    startangle=140,
    colors=plt.cm.Set3.colors,   # nice color palette
    explode=[0.05]*len(fuel_type),  # slight separation for all slices
)
```

```python
company_count.plot(kind='bar')
```

```python
company_vehiclecounts = ds["company"].value_counts()

ax = company_vehiclecounts.plot(kind = 'bar', figsize=(12, 6), color='skyblue')

plt.title("Number of Vehicles per Company")

plt.xlabel("Company")

plt.ylabel("Vehicle Count")

plt.xticks(rotation=45)

#Add count labels on top of bars

for i, value in enumerate (company_vehiclecounts):

    plt.text(i, value + 1, str(value), ha='center', va='bottom', fontsize=9)

plt.tight_layout()
plt.show()
 
```

```python
year_vehiclecounts = ds["year"].value_counts()

ax = year_vehiclecounts.plot(kind = 'bar', figsize=(12, 6), color='skyblue')

plt.title("Number of Vehicles per Year")

plt.xlabel("Year")

plt.ylabel("Vehicle Count")

plt.xticks(rotation=45)

#Add count labels on top of bars

for i, value in enumerate (year_vehiclecounts):

    plt.text(i, value + 1, str(value), ha='center', va='bottom', fontsize=9)

plt.tight_layout()

plt.show()
```

```python
fuel_type_vechicalcount =  ds["fuel_type"].value_counts()
fuel_type_vechicalcount
ax = fuel_type_vechicalcount.plot(kind = 'bar', figsize=(12, 6), color='skyblue')

plt.title("Number of Vehicles on fuel type")

plt.xlabel("Year")

plt.ylabel("Vehicle Count")

plt.xticks(rotation=45)

#Add count labels on top of bars

for i, value in enumerate (fuel_type_vechicalcount):

    plt.text(i, value + 1, str(value), ha='center', va='bottom', fontsize=9)

plt.tight_layout()

plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.swarmplot(
    data=ds,
    x="fuel_type",
    y=None,
    size=6
)
plt.title("Swarm Plot of Vehicles by Fuel Type", fontsize=16)
plt.xlabel("Fuel Type", fontsize=14)
plt.ylabel("Vehicles", fontsize=14)

plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
