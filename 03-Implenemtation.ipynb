```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
ds = pd.read_csv("clean_data.csv")
ds.info()
```

```python
ds.columns
```

```python
X = ds[['name', 'company', 'year', 'kms_driven',
       'fuel_type']]
X
```

```python
y = ds[['Price']]
y
```

```python
from sklearn.compose import make_column_transformer
from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder()
```

```python
ohe.fit(ds[['name', 'company','fuel_type']])
```

```python
ohe.categories_
```

```python
ct =make_column_transformer((OneHotEncoder(categories=ohe.categories_),["name","company","fuel_type"]),remainder="passthrough")
ct
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
```

```python
from sklearn.pipeline import make_pipeline
```

```python
pipe = make_pipeline(ct,reg)
pipe
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y, test_size=0.1,random_state=0)
```

```python
y_train
```

```python
pipe.fit(X_train,y_train)
```

```python
data = pd.DataFrame(data = [["Figo","Ford",2017,25000,"Petrol"]],columns=['name', 'company', 'year', 'kms_driven',
       'fuel_type'])
data

# myinput = pd.DataFrame(data=[["Ford","Figo",2015,50000,"Diesel"]],columns=["company","name","year","kms_driven","fuel_type"])
```

```python
pipe.predict(data)
```

```python
X_test
```

```python
y_pred = pipe.predict(X_test)
y_pred
```

```python
y_pred = pd.DataFrame(data = y_pred,columns=["Result"])
y_pred
```

```python
y_test
```

```python
result  = pd.concat([y_test.reset_index(drop=True),y_pred.reset_index(drop=True)],axis=1)
result
```

```python
from sklearn.metrics import r2_score
score = r2_score(result['Price'],result['Result'])
print(score)
```

```python
new_array = []
for i in range(0,101):
    X_train,X_test,y_train,y_test = train_test_split(X,y, test_size=0.1,random_state=i)
    pipe.fit(X_train,y_train)
    y_pred = pipe.predict(X_test)
    y_pred
    y_pred = pd.DataFrame(data = y_pred,columns=["Result"])
    y_pred
    result  = pd.concat([y_test.reset_index(drop=True),y_pred.reset_index(drop=True)],axis=1)
    result
    score = r2_score(result['Price'],result['Result'])
    new_array.append(score)
```
